# Nutbutter Sales Example

```python
np.random.seed(0)
sales = np.random.randint(0,20,[5,3])
sales
```

```python
weekly_sales = pd.DataFrame(sales,index = ["Mon","Tue","Wed","Thurs","Fri"],columns=["Almond","Peanut","Cashew"])
weekly_sales
```

```python
prices = np.array([10,8,12])
prices
```

```python
butter_prices = pd.DataFrame(prices.reshape(1,3),index = ["Price"],columns = ["Almond","Peanut","Cashew"])
butter_prices
```

```python
total_sales = np.dot(sales,prices)
total_sales.reshape(5,1)
```

```python
weekly_sales["Total Sales"] = total_sales.reshape(5,1)
```

```python
weekly_sales
```

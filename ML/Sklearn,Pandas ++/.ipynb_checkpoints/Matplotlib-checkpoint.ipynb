```python
# Importing new data frame
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease.age.plot.hist(bins = 50);
```

```python
heart_disease.plot.hist(subplots= True,figsize =(10,40));
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
# pyplot method
over_50.plot.scatter(x="age",y="chol",c="target");
```

```python
# OO + Pyplot
fig,ax = plt.subplots(figsize= (10,5))
over_50.plot(kind="scatter",x="age",y= 'chol',c="target",ax=ax);
ax.set_xlim(45,80);
```

```python
# OO Method Only
fig,ax = plt.subplots(figsize=(10,5))
# Plotting data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c = over_50["target"])
ax.set(title="Heart Disease vs Cholestrol Level",xlabel="Age",ylabel="Cholestrol");
# Legend
# It gets legend elements from c in the scatter variable
ax.legend(*scatter.legend_elements(),title = "Target");

# Horizontal line
ax.axhline(over_50["chol"].mean(),linestyle = '--');
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize = (10,10))

# Subplot 1

scatter_1 =ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap='winter')
# Labels 
ax0.set(title="Heart Disease vs Cholestrol Level",xlabel='Age',ylabel='Cholestrol');
# Legends
ax0.legend(*scatter_1.legend_elements(),title="Target")
# Horizontal Line
ax0.axhline(heart_disease["chol"].mean(),linestyle="--");

# Subplot 2

scatter_2 =ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap='winter')
# Labels 
ax1.set(title="Heart Disease vs Thalach Level",xlabel='Age',ylabel='Thalach');
# Legends
ax1.legend(*scatter_2.legend_elements(),title="Target")
# Horizontal Line
ax1.axhline(heart_disease["thalach"].mean(),linestyle="--");
```
